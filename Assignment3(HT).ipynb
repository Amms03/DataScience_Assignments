### Question-3

```python
df= [[50,142,131,70],[435,1523,1356,750]]
```

```python
df
```

```python
stat, p, dof, expected = stats.chi2_contingency(df)
print(f'P-value = {p}')
print(f"Degrees of freedom = {dof}")
print(p)
```

```python
prob= 0.95
```

```python
alpha = 1.0 - prob
print(f"alpha = {alpha}, p_value = {p} ")
if p<= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```

### Question-4

```python
df= [[29,33,31,20],[271,267,269,280]]
```

```python
df
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
import numpy as np
stat, p, dof, expected = stats.chi2_contingency(df)
print('dof=%d' % dof)
print(p)
```

```python
alpha = 0.05
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```
