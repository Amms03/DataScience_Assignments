```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("wine.csv")
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler #normalizing

scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
pca = PCA(n_components=14)
pca_values = pca.fit_transform(scaled_data)
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1,color="red")
```

```python
pca_values[:,0:3]
```

```python
x = pca_values[:,0:1]
y = pca_values[:,1:2]
z = pca_values[:,2:3]
plt.scatter(x,y,z)
```

```python
final_df = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=["pc1","pc2","pc3"]),df[["Type"]]],axis=1)
```

```python
final_df
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
final_df
```

```python
dendrogram = sch.dendrogram(sch.linkage(final_df,method="average"))
```

```python
hc = AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="average")
```

```python
y_hc = hc.fit_predict(final_df)
clusters = pd.DataFrame(y_hc, columns = ["Clusters"])
```

```python
final_df["h_clusterid"] = clusters
```

```python
final_df.sort_values("h_clusterid")
```

```python
final_df.iloc[:,:].groupby(final_df.h_clusterid).mean()
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="average").fit(final_df)
cluster_labels = pd.Series(hc.labels_)
```

```python
df["clusterid"]=cluster_labels
df
```

```python
df.iloc[:,:].groupby(df.clusterid).mean()
```

```python
from sklearn.cluster import KMeans
```

```python
f_df = final_df.iloc[:,0:4]
f_df
```

```python
wcss=[]
for i in range(1,14):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(f_df)
    wcss.append(kmeans.inertia_)
wcss
```

```python
wcss=[]
for i in range(1,14):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(f_df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,14),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
```

```python
# Building Cluster Algorithm

cluster_new = KMeans(3,random_state=42)
cluster_new.fit(f_df)
```

```python
cluster_new.labels_
```

```python
f_df["clusterid_new"]=cluster_new.labels_
```

```python
f_df.groupby("clusterid_new").agg(["mean"]).reset_index()
```
